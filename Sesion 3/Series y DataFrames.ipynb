# Series y DataFrames

Para llamar a Pandas usando las convenciones habituales

```python
import pandas as pd
```

Crearemos una serie a partir de una lista 

```python
lista=[a for a in range(1,101)]
print(lista)
```

```python
pd_series1= pd.Series(lista)
pd_series1
```

Igualmente se puede hacer lo mismo con un arreglo de numpy

```python
import numpy as np
arr_np=np.arange(1,101)
arr_np
```

```python
pd_series2=pd.Series(arr_np)
pd_series2
```

Es posible crear estructuras con listas y diccionarios para después convertirlas en un formato de Pandas.

```python
df=pd.DataFrame({'key':['A','B','C','D','E','F','A','B','C','D','E','F'],
                 'datosA':list(range(1,13,1)),'datosB':np.random.randint(0,10,12)})
df
```

Puede cambiarse el índice principal del DataFrame y que ahora se pueda identifcar cada renglón con una letra

```python
df=df.set_index('key')
df
```

Para usar slicing sobre un objeto de Pandas es necesario hacerlo a través de dos métodos especiales:
iloc y loc

Usando loc

```python
df.loc['F']
```

Usando iloc

```python
df.iloc[2:5]
```
