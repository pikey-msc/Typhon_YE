### 6. Resuelva los siguientes incisos:

### A. Hacer una función que reciba un entero y devuelva en una lista todos los números con los que sea divisible. Por ejemplo, si damos 15, se tendría que regresar 3 y 5

Solución

```python
def Divisores(entero):
    lista_divisores=[]
    for r in range(2,entero):
        if entero%r==0:
            lista_divisores.append(r)
    return(lista_divisores)
```

```python
Divisores(20)
```

Por list comprehension

```python
def Divisores_lc(entero):
    lista_divisores=[r for r in range(2,entero) if entero%r==0]
    return(lista_divisores)
```

```python
Divisores_lc(20)
```

#### B. Encuentre el porcentaje de personas que sean millenials en esta base de datos. Recuerde que el quinto y sexto espacio de izquierda a derecha corresponde al año de nacimiento

Solución

```python
contar=0
for registro in curp:
    if 81<=int(registro[4:6])<=93:
        contar+=1
porcentaje1=100*contar/len(curp)
print(str(round(porcentaje1,2))+'%')
```

```python
porcentaje2=100*len([registro for registro in curp if 81<=int(registro[4:6])<=93])/len(curp)
print(str(round(porcentaje2,2))+'%')
```
